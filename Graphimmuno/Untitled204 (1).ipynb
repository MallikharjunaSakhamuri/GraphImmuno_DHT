```python
import numpy as np
import linecache
```

```python
line_positions = []
line_numbers = []
lines = []

search_text = "M rows = ARNDCQEGHILKMFPSTWYV, cols = ARNDCQEGHILKMFPSTWYV"
with open("aaindex3.txt") as f: 
    for lineno, line in enumerate(f): 
        if search_text in line: 
          line_positions.append(lineno)

for num in line_positions:
  for k in range(num+2,num+22): ## 9 +20 (because of 20 amino acid values)
    line_numbers.append(k) 


line_numbers = np.array(line_numbers).reshape(-1, 20)    

Edge_array = np.zeros((47,20, 20))  
#for anum in range(47):
 # globals()[f"Edge_array_{anum}"] = np.zeros((20, 20))

```

```python
line_numbers.shape
```

```python
k = 0
for matricind in line_numbers:
  i = 0
  for seq in matricind:
    x = linecache.getline(r"aaindex3.txt", seq).strip()

    for j in range(20):
      if (j < (len(x.split()))):
        if(x.split()[j] == "NA"):
          Edge_array[k][i][j] = 0 
          
        else:
          Edge_array[k][i][j] = float(x.split()[j])    

   

      #print(i,j)
    i = i+1
  k = k+1
  
```

```python
for k in range(Edge_array.shape[0]):
  for i in range(Edge_array.shape[1]):
    for j in range(Edge_array.shape[2]):
      Edge_array[k][i][j] = Edge_array[k][j][i]        
```

```python
for i in range(3):
  print(Edge_array[i])


#print((Edge_array.shape))  
```

```python
from sklearn.preprocessing import normalize

Edge_array_min = Edge_array.min(axis=(1, 2), keepdims=True)
Edge_array_max = Edge_array.max(axis=(1, 2), keepdims=True)

Edge_array = (Edge_array - Edge_array_min)/(Edge_array_max-Edge_array_min)
```

```python
for i in range(3):
  print(Edge_array[i])
```

```python
Edge_array_mean = Edge_array.mean(axis=(1,2))
Edge_array_mean
```

```python
Edge_Weight_amino = np.zeros((20,47))
Edge_Weight_mean = np.zeros((20,1))


for i in range(Edge_array.shape[1]):
  for j in range(Edge_array.shape[2]):
    for k in range(Edge_array.shape[0]):
      Edge_Weight_amino[j][k] = Edge_array[k][i][j]

    Edge_Weight_mean[i] = Edge_Weight_amino[i].mean()  

#print(Edge_Weight_amino)
print(Edge_Weight_mean)

    
```
